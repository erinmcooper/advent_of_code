```python
def compute_score(input_file="input_test.txt"):
    
    with open(input_file) as f:
        data = f.readlines()
        
    calls = data[0].strip("\n").split(',')
    boards = {}
    num_boards = (len(data)-1)/6
    for board in range(int(num_boards)):
        boards[board+1] = np.asarray([row.split("\n")[0].split() for row in data[2+6*board:][:5]])
        
    fastest = len(calls)
    slowest = 0
    bingoed_boards = []
    for idx in range(5,len(calls)+1):
        for board in boards:
            if board in bingoed_boards:   
                continue
            else:
                for row in boards[board]:
                    row_bingo = set(row).issubset(set(calls[:idx]))
                    if row_bingo:
                        bingoed_boards.append(board)
                        if idx < fastest:
                            fastest = idx
                            winning_board = board
                            win_type = "row"
                        elif idx > slowest:
                            slowest = idx
                            losing_board = board
                            lose_type = "row"
                for col in boards[board].transpose():
                    col_bingo = set(col).issubset(set(calls[:idx]))
                    if col_bingo:
                        bingoed_boards.append(board)
                        if idx < fastest:
                            fastest = idx
                            winning_board = board
                            win_type = "col"
                        elif idx > slowest:
                            slowest = idx
                            losing_board = board
                            lose_type = "col"
    print("WIN: ", winning_board, win_type, fastest, calls[:fastest])
    print("LOSE: ", losing_board, lose_type, slowest, calls[:slowest])
    nums_not_called_in_winning_board = []
    for row in boards[winning_board]:
        for col in row:
            if col not in calls[:fastest]:
                nums_not_called_in_winning_board.append(int(col))
    nums_not_called_in_losing_board = []
    for row in boards[losing_board]:
        for col in row:
            if col not in calls[:slowest]:
                nums_not_called_in_losing_board.append(int(col))
    print("WIN: ", sum(nums_not_called_in_winning_board)*int(calls[fastest-1]))
    print(sum(nums_not_called_in_winning_board), calls[fastest-1])
    print("LOSE: ", sum(nums_not_called_in_losing_board)*int(calls[slowest-1]))
    print(sum(nums_not_called_in_losing_board), calls[slowest-1])
    
    return calls,boards,bingoed_boards
```

```python
calls,boards,bingoed_boards = compute_score()
```

```python
calls,boards,bingoed_boards = compute_score("input.txt")
```
