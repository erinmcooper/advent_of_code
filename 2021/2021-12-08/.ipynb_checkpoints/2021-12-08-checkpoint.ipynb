```python
import pandas as pd
import numpy as np
import math
```

```python
def count_1478s(input_file="test_input.txt"):
    
    with open(input_file, 'r') as f:
        data = f.readlines()
        
    output = [line.split('|')[-1].split() for line in data]
    
    lengths = {1:2, 4:4, 7:3, 8:7}
    counts = {1:0, 4:0, 7:0, 8:0}
    for line in output:
        for num in lengths:
            counts[num] += sum([1 for word in line if len(word)==lengths[num]])
    
    print("1478s in output: ", sum(counts.values()))
    
    return counts
```

```python
def create_cipher(line):

    inputs = line.split("|")[0].split()
    
    unique_lengths = {2:1, 4:4, 3:7, 7:8}
    input_digits = {inputs[i]: np.NaN for i in range(len(inputs))}
    cipher = {i: '' for i in range(10)}
    for digit in input_digits:
        if len(digit) in unique_lengths:
            input_digits[digit] = unique_lengths[len(digit)]
            
    reverse_dict = {input_digits[letters]: letters for letters in input_digits if math.isfinite(input_digits[letters])}
    cipher = reverse_dict.copy()
    for code in input_digits:
        if len(code) == 6:
            if sum([cipher[1][i] in code for i in range(len(cipher[1]))])==1:
                cipher[6] = code
                input_digits[code] = 6
            elif sum([cipher[4][i] in code for i in range(len(cipher[4]))])==4:
                cipher[9] = code
                input_digits[code] = 9
            else:
                cipher[0] = code
                input_digits[code] = 0
    for code in input_digits:
        if len(code) == 5:
            if (cipher[1][0] in code) and (cipher[1][1] in code):
                cipher[3] = code
                input_digits[code] = 3
            elif sum([code[i] in cipher[6] for i in range(len(code))])==5:
                cipher[5] = code
                input_digits[code] = 5
            else:
                cipher[2] = code
                input_digits[code] = 2
                
    return cipher
```

```python
def cipher_output(line, cipher):
    
    outputs = line.split("|")[1].split()
    
    ciphered_outputs = ''
    for output in outputs:
        for num in cipher:
            if sorted(output) == sorted(cipher[num]):
                ciphered_outputs += repr(num)
                
    return int(ciphered_outputs)
```

```python
def cipher_and_sum(input_file="test_input.txt"):
    
    with open(input_file, 'r') as f:
        data = f.readlines()
    
    data = [line.strip("\n") for line in data]
    
    nums_to_sum = []
    for line in data:
        cipher = create_cipher(line)
        ciphered_outputs = cipher_output(line, cipher)
        print(line,"\t", ciphered_outputs)
        nums_to_sum.append(ciphered_outputs)
        
    print()
    print("sum of ciphered outputs: ", sum(nums_to_sum))
    
    return
```
