```python
import numpy as np
from itertools import cycle
```

```python
def take_a_turn(player, roll, space, score, board, die_rolls):
    
    advance = [next(roll) for i in range(3)]
    for move in range(sum(advance)):
        current = next(board)
    score += current
    die_rolls += 3
    
    print(f"{player} rolls {advance} and moves {sum(advance)} to space {current} for a total score of {score}")
    
    return roll, current, score, die_rolls
```

```python
def compute_the_thing(input_file="test_input.txt"):
    
    with open(input_file, "r") as f:
        data = f.readlines()
        
    p1_space = int(data[0].split()[-1])
    p2_space = int(data[1].split()[-1])
    
    print(f"Player 1 starting position: {p1_space}")
    print(f"Player 2 starting position: {p2_space}")
    
    die = list(range(1,101))
    roll = cycle(die)
    
    spaces = list(range(1,11))
    p1_board = cycle(spaces)
    [next(p1_board) for i in range(p1_space)]
    p2_board = cycle(spaces)
    [next(p2_board) for i in range(p2_space)]
    
    p1_score = 0; p2_score = 0; die_rolls = 0
    while p1_score < 1000 or p2_score < 1000:
        
        # p1 takes a turn
        roll, p1_space, p1_score, die_rolls = take_a_turn("player 1", roll, p1_space, p1_score, p1_board, die_rolls)
        
        if p1_score >= 1000:
            print(f"player 1 wins! ({p1_score}) player 2 score ({p2_score}) * die_rolls ({die_rolls}) = {p2_score * die_rolls}")
            return p1_score, p2_score, die_rolls
        
        # p2 takes a turn
        roll, p2_space, p2_score, die_rolls = take_a_turn("player 2", roll, p2_space, p2_score, p2_board, die_rolls)
        
        if p2_score >= 1000:
            print(f"player 1 wins! ({p2_score}) player 1 score ({p1_score}) * die_rolls ({die_rolls}) = {p1_score * die_rolls}")
            return p1_score, p2_score, die_rolls
    
```

```python
p1_score, p2_score, die_rolls = compute_the_thing()
```

```python
def take_a_turn_dirac(player, space, score, board, die_rolls):
    
    possible_moves = []
    for move in range(3,10):
        if space+move > 10:
            possible_moves.append((space+move)%10) 
        elif space+move <= 10:
            possible_moves.append(space+move)
    space = min(possible_moves)
    score += min(possible_moves)
    die_rolls += 3
    
    print(f"{player} moves to space {space} for a total score of {score} and total multiverse count of {3 ** die_rolls}")
    
    return space, score, die_rolls
```

```python
def compute_the_multiverse(input_file="test_input.txt"):
    
    with open(input_file, "r") as f:
        data = f.readlines()
        
    p1_space = int(data[0].split()[-1])
    p2_space = int(data[1].split()[-1])
    
    print(f"Player 1 starting position: {p1_space}")
    print(f"Player 2 starting position: {p2_space}")
    
    spaces = list(range(1,11))
    p1_board = cycle(spaces)
    [next(p1_board) for i in range(p1_space)]
    p2_board = cycle(spaces)
    [next(p2_board) for i in range(p2_space)]
    
    p1_score = 0; p2_score = 0
    p1_die_rolls = 0; p2_die_rolls = 0;
    while p1_score < 21 or p2_score < 21:
        
        # p1 takes a turn
        p1_space, p1_score, p1_die_rolls = take_a_turn_dirac("player 1", p1_space, p1_score, p1_board, p1_die_rolls)
        
        if p1_score >= 21:
            print(f"player 1 wins! multiverses 3 ** die_rolls = {3 ** p1_die_rolls}")
            return p1_score, p2_score, p1_die_rolls
        
        # p2 takes a turn
        p2_space, p2_score, p2_die_rolls = take_a_turn_dirac("player 2", p2_space, p2_score, p2_board, p2_die_rolls)
        
        if p2_score >= 21:
            print(f"player 2 wins! multiverses 3 ** die_rolls = {3 ** p2_die_rolls}")
            return p1_score, p2_score, p2_die_rolls
    
```

```python
p1_score, p2_score, die_rolls = compute_the_multiverse()
```
